```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cobweb_cubic_root(x0=0, C=10, tolerance=0.0001, max_iter=10):
    fig, ax = plt.subplots(figsize=(10, 8)) # draw a plot
    x_range = np.linspace(0.1, 8, 500) #create x values
    y_line = x_range  # y = x
    y_curve = 0.5 * (x_range + C / (x_range**2))  #cubic root formula
    
    # Plot the curves
    ax.plot(x_range, y_line, 'orange', linewidth=2, label='y = x') #plot original y=x curve
    ax.plot(x_range, y_curve, 'blue', linewidth=2, label=f'y = 1/2(x + {C}/x)') #plot original cubic root curve
    
    ax.set_xlim(0, 8)
    ax.set_ylim(0, 8)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(f'Geometric Analysis for {C}')
    ax.grid(True, alpha=0.3)
    
    # Perform iterations
    x = x0
    path_x = [x] #for x start with x0
    path_y = [0] #for y start with 0
    
    for i in range(max_iter): #for a number of iterations
        #go vertical
        print(f'iteration {i}')
        y_new = 0.5 * (x + C / (x**2)) #find a new y coordinate
        print(f'y(x) is {y_new}')
        path_x.append(x) #append x vaues
        print(f'adding x coord {x}')
        path_y.append(y_new) #append y values
        print(f'start at y = {y_new}')
        
        #go horizontal 
        x_new = y_new #path from cubic curve to y=x curve
        path_x.append(x_new) #append x coord
        print(f'from y go to y=x and record x coord {x_new}')
        path_y.append(y_new) #append y coord
        print(f'record y coord of this horizontal step {y_new}')
        
        print(f"x{i} = {x:.6f} -> x{i+1} = {x_new:.6f}")
        
        if abs(x_new - x) < tolerance:
            break
        x = x_new
    
    # Draw the path
    ax.plot(path_x, path_y, 'g-', linewidth=1, alpha=0.7, marker='o', markersize=3)
    ax.plot(x, x, 'r*', markersize=12, label=f'Converged: x = {x:.6f}')
    
    # Mark starting point
    ax.plot(x0, 0, 'go', markersize=8, label=f'Start: x = {x0}')
    
    ax.legend()
    plt.show()
    return x
result = cobweb_cubic_root(x0=1, C=10)
```

## Probelm 2: Alternative way of finding cubic root

```python
def cobweb_cubic_root_2(x0=1, C=10, tolerance=0.0001, max_iter=20):
    fig, ax = plt.subplots(figsize=(10, 8))
    x_range = np.linspace(0.1, 8, 500)
    
    y_cubic = 0.5 * (x_range + C / (x_range**2))  # Original Babylonian for cubic root
    y_alternative = 0.5 * (x_range + (C / x_range)**(1/2))  # Alternative with square root
    
 
    ax.plot(x_range, y_cubic, 'blue', linewidth=2, label=f'Original: y = 1/2(x + {C}/x)')
    ax.plot(x_range, y_alternative, 'orange', linewidth=2, label=f'Alternative: y = 1/2(x + ({C}/x))')
    
    ax.set_xlim(0, 8)
    ax.set_ylim(0, 8)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(f'Alternative Method for Finding {C}')
    ax.grid(True, alpha=0.3)

    x = x0
    path_x = [x]
    path_y = [0]
    
    for i in range(max_iter):
        print(f'iteration {i}')
        
        y_new = 0.5 * (x + (C / x)**(1/2)) 
        print(f'y = {y_new:.6f}')
        path_x.append(x)
        print(f'record x coord for vertical line {x}')
        path_y.append(y_new)
        print(f'record y coord for vertical line {y_new}')
        x_new = y_new  
        path_x.append(x_new)
        path_y.append(y_new)
        print(f'record x and y coordinates for horizontal line {x_new, y_new}')
        print(f"x{i} = {x:.6f} -> x{i+1} = {x_new:.6f}")
        
        if abs(x_new - x) < tolerance:
            break
        x = x_new

    ax.plot(path_x, path_y, 'g-', linewidth=1.5, alpha=0.7, marker='o', markersize=4)
    
    ax.plot(x, x, 'r*', markersize=12, label=f'Converged: x = {x:.6f}')
    
    ax.plot(x0, 0, 'go', markersize=8, label=f'Start: x = {x0}')
    
    ax.legend()
    plt.show()
    return x


result = cobweb_cubic_root_2(x0=3, C=64)
print(f"\nFinal result from alternative method: {result:.8f}")
```

As you can see, the alternative method for finding cubic root of 64 does it faster (a.k.a with fewer iterations) and finds the result that is very close to accurate (3.999 vs 4.00)

## Graphical analysis of logistic map

```python
def cobweb_logistic(x0=0.2, r=3.2, max_iter=30):
    fig, ax = plt.subplots(figsize=(10, 8))
    x_range = np.linspace(0, 1, 800)
    y_line = x_range
    y_curve = r * x_range * (1 - x_range)

    ax.plot(x_range, y_line, linewidth=2, label="y = x")
    ax.plot(x_range, y_curve, linewidth=2, label=f"y = r x (1 - x), r={r}")

    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(f'Cobweb Diagram for Logistic Map (r={r})')
    ax.grid(True, alpha=0.3)

    xn = x0
    path_x = [x0]
    path_y = [0.0]

    for i in range(max_iter):
        yn = r * xn * (1 - xn)
        path_x.append(xn) #coord of vert move in x direction
        path_y.append(yn) #coord of vert move in y direction

    
        path_x.append(yn) #coord of horizontal move in x dir
        path_y.append(yn) #coor of horixontal move in x dir
        xn = yn 

    ax.plot(path_x, path_y, '-', linewidth=1, alpha=0.8, marker='o', markersize=3, label='Cobweb path')
    ax.plot(xn, xn, marker='*', markersize=12, label=f'Last iterate: x  {xn:.6f}')

    ax.plot(x0, 0, marker='o', markersize=8, label=f'Start: x = {x0}')

    ax.legend()
    plt.show()
    return xn

# initial = 0.2, r = 3.25
trial1 = cobweb_logistic(x0=0.5, r=3.25, max_iter=50)
```
